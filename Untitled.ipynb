```python
def remove_pps_outliers(data_announces):
    data_announces_out=pd.DataFrame()
    for key, subdf in data_announces.groupby('location_area'):
        m= np.mean(subdf.price_per_unit)
        st=np.std(subdf.price_per_unit)  #standard dev.
        reduce_data_announces=subdf[(subdf.price_per_unit>(m-st)) & (subdf.price_per_unit<=(m+st))]
        data_announces_out=pd.concat([data_announces_out, reduce_data_announces],ignore_index=True)
    return data_announces_out

data_announces3=remove_pps_outliers(data_with_full_price)
data_announces3.shape
```

# Impartire in subseturi in functie de locatie

```python
# Lista cu locaÈ›iile unice care au cel puÈ›in 20 de anunÈ›uri
locations = data_announces3_clean['location_area'].value_counts()[data_announces3_clean['location_area'].value_counts() >= 20].index.tolist()

# DicÈ›ionar pentru a stoca modelele È™i coeficienÈ›ii pentru fiecare locaÈ›ie
models = {}
coefficients = {}

# Parcurgem fiecare locatie si cream un model separat pentru ea
for location in locations:
    # Filtram datele pentru locatia curenta
    subset = data_announces3_clean[data_announces3_clean['location_area'] == location]
    
    # Verificam daca avem cel putin 20 de anunturi in locatie
    if subset.shape[0] >= 20:
        # Cream setul de date
        X = subset.drop(['price', 'location_area'], axis=1)
        y = subset['price']

        # AdaugÄƒm o coloanÄƒ de 1 la Ã®nceputul setului de date
        X = np.c_[np.ones(X.shape[0]), X]

        # SeparÄƒm setul de date Ã®n setul de antrenare È™i cel de testare
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)

        # IniÈ›ializÄƒm modelul de regresie liniarÄƒ cu o constantÄƒ
        model = LinearRegression(fit_intercept=True)

        # AntrenÄƒm modelul
        model.fit(X_train, y_train)

        # AdÄƒugÄƒm modelul È™i coeficienÈ›ii Ã®n dicÈ›ionare
        models[location] = model
        coefficients[location] = model.coef_

        # EfectuÄƒm predicÈ›ii pe setul de testare
        y_pred = model.predict(X_test)

        # AfiÈ™Äƒm tabelul cu valoarea realÄƒ È™i cea prezisÄƒ
        print("LocaÈ›ia: {}".format(location))
        print(pd.DataFrame({'Valoare reala': y_test, 'Valoare prezisa': y_pred}))

        # AfiÈ™Äƒm graficul pentru locaÈ›ia curentÄƒ
        plt.scatter(y_test, y_pred, color='blue', label='Valoare prezisa')
        plt.plot(y_test, y_test, color='red', label='Valoare reala')
        plt.xlabel("Valoare reala")
        plt.ylabel("Valoare prezisa")
        plt.title("LocaÈ›ia: {}".format(location))
        plt.legend()
        plt.show()
    else:
        print("Nu exista suficiente date pentru locatia: {}".format(location))



    

```
